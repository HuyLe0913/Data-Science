```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
population = pd.read_csv(r"C:\Users\dmin\HUST\20241\DataScience\cleaned_data.csv")

population.head(20)
```

```python
print(f"Số cột: {len(population.index)}")
print(f'Số quốc gia: {len(set(population["Country"].values))}')

for nation in set(population["Country"].values):
    matching_rows = population[population["Country"] == nation]
    count = len(matching_rows)
    print(f'{nation} có {count} hàng dữ liệu')
    
print(f'Số vùng lãnh thổ riêng biệt: {len(set(population["SDG Region"].values))}')
print(set(population["SDG Region"].values))
```

```python
mean_population = np.mean(population["Population"].values)
max_population = np.max(population["Population"].values)
min_population = np.min(population["Population"].values)
print(f'Mean, max and min population :{mean_population,max_population,min_population}')

mean_old_age_dependency_ratio = np.mean(population["Retirement Age Dependency Ratio"].values) 
max_old_age_dependency_ratio = np.max(population["Retirement Age Dependency Ratio"].values)
min_old_age_dependency_ratio = np.min(population["Retirement Age Dependency Ratio"].values)

print(f'Min, max and mean retirement age dependency ratio :{min_old_age_dependency_ratio,max_old_age_dependency_ratio,mean_old_age_dependency_ratio}')

mean_young_dependency_ratio = np.mean(population["Young Age Dependency Ratio"].values)
max_young_dependency_ratio = np.max(population["Young Age Dependency Ratio"].values)
min_young_dependency_ratio = np.min(population["Young Age Dependency Ratio"].values)
print(f'Mean, max and min young dependency ratio : {mean_young_dependency_ratio,max_young_dependency_ratio,min_young_dependency_ratio}')

mean_area = np.mean(population["Total Area (sq km)"].values)
max_area = np.max(population["Total Area (sq km)"].values)
min_area = np.min(population["Total Area (sq km)"].values)
print(f'Mean, max and min area : {mean_area,max_area,min_area}')

mean_gdp = np.mean(population["GDP_per_Capita"].values)
max_gdp = np.max(population["GDP_per_Capita"].values)
min_gdp = np.min(population["GDP_per_Capita"].values)
print(f'Mean, max and min GDP : {mean_gdp,max_gdp,min_gdp}')


mean_infant_mortality = np.mean(population["Infant mortality per 1000 live births"].values)
max_infant_mortality = np.max(population["Infant mortality per 1000 live births"].values)
min_infant_mortality = np.min(population["Infant mortality per 1000 live births"].values)
print(f'Mean, max and min infant mortality per 1000 live births : {mean_infant_mortality,max_infant_mortality,min_infant_mortality}')

mean_bmi = round(np.mean(population["BMI"].values),1)
max_bmi = np.max(population["BMI"].values)
min_bmi = np.min(population["BMI"].values)
print(f'Mean, max and min BMI : {mean_bmi,max_bmi,min_bmi}')

mean_male_ratio = np.mean(population["Male Population Ratio"].values)
max_male_ratio = np.max(population["Male Population Ratio"].values)
min_male_ratio = np.min(population["Male Population Ratio"].values)
print(f'Mean, max and min male ratio : {mean_male_ratio,max_male_ratio,min_male_ratio}')

mean_urban_population_ratio = np.mean(population["Urban Population Ratio"].values)
max_urban_population_ratio = np.max(population["Urban Population Ratio"].values)
min_urban_population_ratio = np.min(population["Urban Population Ratio"].values)
print(f'Mean, max and min urban population ratio : {mean_urban_population_ratio,max_urban_population_ratio,min_urban_population_ratio}')
```

```python
sgd_region_map = {
    'Central and Southern Asia': 1,
    'Europe and Northern America': 2,
    'Australia and New Zealand': 3,
    'Latin America and the Caribbean': 4,
    'Sub-Saharan Africa': 5,
    'Northern Africa and Western Asia': 6,
    'Oceania': 7,
    'Eastern and South-Eastern Asia': 8
}
num_regions = len(sgd_region_map)

def one_hot_encode_region(region):
    index = sgd_region_map[region] - 1 
    one_hot_vector = np.zeros(num_regions)
    one_hot_vector[index] = 1
    return one_hot_vector

feature_columns = [
    'Young Age Dependency Ratio', 'Retirement Age Dependency Ratio', 'Population',
    'Male Population Ratio', 'Urban Population Ratio', 'Total Area (sq km)',
    'GDP_per_Capita', 'Infant mortality per 1000 live births', 'BMI'
]
label_column = 'Life Expectancy'


precomputed_stats = {
    "Population": (mean_population,max_population,min_population),
    "Retirement Age Dependency Ratio": (mean_old_age_dependency_ratio,max_old_age_dependency_ratio,min_old_age_dependency_ratio),
    "Young Age Dependency Ratio": (mean_young_dependency_ratio,max_young_dependency_ratio,min_young_dependency_ratio),
    "Total Area (sq km)": (mean_area,max_area,min_area),
    "GDP_per_Capita": (mean_gdp,max_gdp,min_gdp),
    "Infant mortality per 1000 live births": (mean_infant_mortality,max_infant_mortality,min_infant_mortality),
    "BMI": (mean_bmi,max_bmi,min_bmi),
    "Male Population Ratio": (mean_male_ratio,max_male_ratio,min_male_ratio),
    "Urban Population Ratio": (mean_urban_population_ratio,max_urban_population_ratio,min_urban_population_ratio)
}


def normalize_column(col, column_name):
    
    mean, max_val, min_val = precomputed_stats[column_name]
    
    return (col - mean) / (max_val - min_val)

train_data = []
validation_data = []

for country, group in population.groupby('Country'):
    group = group.sort_values(by='Year')

    split_index = int(0.8 * len(group))
    train_group = group.iloc[:split_index]
    validation_group = group.iloc[split_index:]

    train_features = []
    validation_features = []
    for col in feature_columns:
        if col in precomputed_stats:
            train_norm = normalize_column(train_group[col], col)
            train_features.append(train_norm)
            validation_features.append(normalize_column(validation_group[col], col))

    train_sdg_region = np.array([one_hot_encode_region(region) for region in train_group['SDG Region']])
    validation_sdg_region = np.array([one_hot_encode_region(region) for region in validation_group['SDG Region']])

    train_features = np.column_stack(train_features + [train_sdg_region])
    validation_features = np.column_stack(validation_features + [validation_sdg_region])

    train_labels = train_group[label_column].to_numpy()
    validation_labels = validation_group[label_column].to_numpy()

    train_data.append(np.column_stack((train_features, train_labels)))
    validation_data.append(np.column_stack((validation_features, validation_labels)))

train_array = np.vstack(train_data)
validation_array = np.vstack(validation_data)

print(f'Train array: {train_array.shape}')
print(f'Validation array: {validation_array.shape}')
print(train_array[0])
```

# Create training set and testing set

```python
import torch
from torch.utils.data import DataLoader, TensorDataset


train_features = []
train_labels = []

for array in train_array:
    train_features.append(array[:-1])  
    train_labels.append(array[-1])     

test_features = []
test_labels = []

for array in validation_array:
    test_features.append(array[:-1])  
    test_labels.append(array[-1])


train_features_tensor = torch.tensor(train_features, dtype=torch.float32)
train_labels_tensor = torch.tensor(train_labels, dtype=torch.float32).unsqueeze(1)  

test_features_tensor = torch.tensor(test_features, dtype=torch.float32)
test_labels_tensor = torch.tensor(test_labels, dtype=torch.float32).unsqueeze(1)    


train_dataset = TensorDataset(train_features_tensor, train_labels_tensor)
test_dataset = TensorDataset(test_features_tensor, test_labels_tensor)


batch_size = 512

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)  
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)   
```

# Declare and training the model

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import copy
import matplotlib.pyplot as plt


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()
        
        self.MLP = nn.Sequential(
            nn.Linear(17,32),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(32,48),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(48,1)
        )

    def forward(self, x):
                    
        return self.MLP(x)


model = NeuralNet().to(device)
criterion = nn.MSELoss()  

optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9, weight_decay=0.0005)

scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=100, verbose=True)

def clip_gradient(optimizer, grad_clip):
    for group in optimizer.param_groups:
        for param in group['params']:
            if param.grad is not None:
                param.grad.data.clamp_(-grad_clip, grad_clip)




def test_the_model(model, test_loader, criterion):
    model.eval()  
    total_loss = 0.0

    
    with torch.no_grad():
        for features, labels in test_loader:
            features = features.to(device)
            labels = labels.to(device)

            outputs = model(features)

            
            loss = criterion(outputs, labels)
            total_loss += loss.item()

    avg_loss = total_loss / len(test_loader)
    print(f"Test Loss: {avg_loss:.4f}")
    return avg_loss


def train_the_model(model, train_loader, test_loader, criterion, optimizer, scheduler, num_epochs=5, grad_clip=1.0):
    train_losses = []
    test_losses = []
    best_model = None
    min_test_loss = float('inf')

    for epoch in range(num_epochs):
        model.train() 
        running_loss = 0.0

        for i, (features, labels) in enumerate(train_loader):
            features = features.to(device)
            labels = labels.to(device)

            optimizer.zero_grad()
            outputs = model(features)
            loss = criterion(outputs, labels)  
            loss.backward()

            
            clip_gradient(optimizer, grad_clip)

            optimizer.step()

            running_loss += loss.item()

        train_loss = running_loss / len(train_loader)
        train_losses.append(train_loss)

        
        test_loss = test_the_model(model, test_loader, criterion)
        test_losses.append(test_loss)

        
        scheduler.step(test_loss)

        
        if test_loss < min_test_loss:
            best_model = copy.deepcopy(model)
            min_test_loss = test_loss
            print(f"Saving best model with Test Loss: {test_loss:.4f}")

        print(f"Epoch {epoch + 1}/{num_epochs}, "
              f"Train Loss: {train_loss:.4f}, "
              f"Test Loss: {test_loss:.4f}")
    print(f'Best training loss after {num_epochs} epochs is: {min(train_losses)}')
    print(f'Best validation loss after {num_epochs} epochs is: {min_test_loss}')
    plt.plot(train_losses, label='Train Loss')
    plt.plot(test_losses, label='Test Loss')
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training and Testing Loss Over Epochs")
    plt.legend()
    plt.show()

    return best_model

best_model = train_the_model(model= model,train_loader= train_loader, test_loader= test_loader, criterion= criterion, optimizer= optimizer, scheduler= scheduler,num_epochs= 1000)
```
